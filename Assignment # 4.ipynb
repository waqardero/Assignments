```python
#Assignment 4.5

import random

def guess_num():
  
    for i in range(3):
        rand = random.randrange(1, 30)
        #print(hidden)
 
        guess = int(input("Please enter your guess: "))
     
        if guess == rand:
            print("Congrats ! U Guessed the Right Number")
        elif guess < rand:
            print("Your guess is too low")
        else:
            print("Your guess is too high")
            
guess_num()            
```

```python
questions = [
    
    {
    "S.no" : 1,
    "question" : "What is Capital City of Pakistan",
    "options" : ['Islamabad', "Karachi","Hyderabad"],
    "answer" :"islamabad"
    },
    
    {
    "S.no" : 2,
    "question" : "What is your City name" ,
    "options" :[ 'Islamabad', "Karachi","Hyderabad"],
    "answer" : "hyderabad"
    },
    
    {
    "S.no" : 3,
    "question" : "What is Country Name?",
    "options" : ['America', "India","Pakistan"],
    "answer" : "pakistan"
    },
    
    {
    "S.no" : 4,
    "question" : "Who is the founder of computer",
    "options" : ["charles babbage","bill gates","Jeck ma"],
    "answer" :"charles babbage"
    }
    
    
]

score = 0

for question in questions:
    data = str(question["S.no"]) + " : "  + question["question"] + "\n" +  question["options"][0] + "\n" + question["options"][1] + "\n" + question["options"][2] + "\n"  
    answer = input(data)
    if answer == question["answer"]:
        score += 1
        
per = ((score *25) /100) * 100         
print(" Your Score is " , str(score) )         
print("\n Your Percentage is " ,per  )        
    
```
